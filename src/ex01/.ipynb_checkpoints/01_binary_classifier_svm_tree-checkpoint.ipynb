## 1. Preprocessing and logreg visualization

Read the file `am_pm.csv` to a dataframe from the previous exercise.

1. Draw a plot where the x-axis is `am`, the y-axis is `pm`, dots are the days, color depends on the target.
2. Add the decision boundary of logistic regression to the plot.
3. Draw the same plot (with the boundary), but the color should depend this time on the predictions.

Now it should be clear for you how the logistic regression works.

```python
df = pd.read_csv("../../data/am_pm.csv")
df.head()
```

```python
plt.scatter(df['am'], df['pm'], c=df['target_encoded'], cmap='viridis', edgecolors='k')
plt.xlabel('am')
plt.ylabel('pm')
plt.title('Scatter plot of am vs pm with target color')
plt.colorbar(label='Target (0 - weekend, 1 - working_day)')
plt.show()
```

```python
X = df[['am', 'pm']].values 
y = df['target_encoded'].values
model = LogisticRegression(random_state=21, fit_intercept=False)
model.fit(X, y)
```

```python
xx, yy = np.meshgrid(np.arange(df['am'].min(), df['am'].max(), 0.1),
                     np.arange(df['pm'].min(), df['pm'].max(), 0.1))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
```

```python
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(df['am'], df['pm'], c=df['target_encoded'], cmap='coolwarm', edgecolors='k')
plt.xlabel('am')
plt.ylabel('pm')
plt.title('Logistic Regression Decision Boundary')
plt.colorbar(label='Target')
plt.show()
```

```python
predictions = model.predict(X)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(df['am'], df['pm'], c=predictions, cmap='coolwarm', edgecolors='k')
plt.xlabel('am')
plt.ylabel('pm')
plt.title('Logistic Regression Predictions')
plt.colorbar(label='Predictions')
plt.show()
```

## 2. Standardizing and logreg

Linear models can be sensitive to the scale of your variables. You make it easier for them to find the optimal solution when you scale your features.

1. Apply `StandardScaler` to `X` and train logistic regression again with the same parameters.
2. Calculate `accuracy` for the new model. Did it get better than a naive classifier with the most popular class?
3. Draw the plots that were described above but for the new model.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model_scaled = LogisticRegression(random_state=21, fit_intercept=False)
model_scaled.fit(X_scaled, y)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred_scaled = model_scaled.predict(X_scaled)

accuracy_scaled = accuracy_score(y, y_pred_scaled)
accuracy_scaled
```

```python
y_series = pd.Series(y)
naive_accuracy = y_series.value_counts(normalize=True).max()
naive_accuracy
```

```python
xx, yy = np.meshgrid(
    np.arange(X_scaled[:, 0].min() - 0.5, X_scaled[:, 0].max() + 0.5, 0.1),
    np.arange(X_scaled[:, 1].min() - 0.5, X_scaled[:, 1].max() + 0.5, 0.1)
)


grid = np.c_[xx.ravel(), yy.ravel()]


probs = model_scaled.predict_proba(grid)[:, 1].reshape(xx.shape)

plt.contour(xx, yy, probs, levels=[0.5], cmap="Greys", vmin=0, vmax=1, linestyles="--")
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap="coolwarm", edgecolor="k")
plt.xlabel("am (scaled)")
plt.ylabel("pm (scaled)")
plt.title("Logistic Regression Decision Boundary (Scaled)")
plt.show()

```

## 3. SVM

1. Apply `SVC` model to the scaled dataframe using parameters `probability=True`, `random_state=21`.
2. Calculate `accuracy`.
3. Try different kernels, find the best in terms of accuracy.
4. Draw both plots again with the decisions boundary to see how this algorithm works.

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc_model = SVC(kernel='linear', probability=True, random_state=21)
svc_model.fit(X_scaled, y)
```

```python
y_pred = svc_model.predict(X_scaled)
accuracy = accuracy_score(y, y_pred)
accuracy
```

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernels:
    svc_model = SVC(kernel=kernel, probability=True, random_state=21)
    svc_model.fit(X_scaled, y)
    accuracy = svc_model.score(X_scaled, y)
    print(f"Accuracy with {kernel} kernel: {accuracy:.3f}")
```

```python
rbfsvc = SVC(kernel="rbf")
rbfsvc.fit(X_scaled, y)
```

```python
xx, yy = np.meshgrid(
    np.linspace(X_scaled[:, 0].min() - 0.5, X_scaled[:, 0].max() + 0.5, 200),
    np.linspace(X_scaled[:, 1].min() - 0.5, X_scaled[:, 1].max() + 0.5, 200)
)
grid = np.c_[xx.ravel(), yy.ravel()]
```

```python
probs = rbfsvc.decision_function(grid).reshape(xx.shape)
```

```python
y_pred = rbfsvc.predict(X_scaled)
```

```python
plt.figure(figsize=(5, 5))
plt.contourf(xx, yy, probs, levels=50, cmap="coolwarm", alpha=0.8)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_pred, cmap="coolwarm", edgecolor="k")
plt.title("SVC result using rbf kernel")
plt.xlabel("am (scaled)")
plt.ylabel("pm (scaled)")
plt.colorbar(label="Decision function value")
plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.contourf(xx, yy, probs, levels=50, cmap="coolwarm", alpha=0.8)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap="coolwarm", edgecolor="k")
plt.title("SVC result using rbf kernel")
plt.xlabel("am (scaled)")
plt.ylabel("pm (scaled)")
plt.colorbar(label="Decision function value")
plt.show()
```

## 4. Decision tree

1. Apply `DecisionTreeClassifier` to the scaled dataframe using parameters `max_depth=4`, `random_state=42`.
2. Calculate accuracy.
3. Try different values of `max_depth`.
4. Draw both plots again with the decisions boundary to see how this algorithm works.
5. Using method `.plot_tree()` visualize the decision tree itself. It is another way of thinking about how the algorithm works.
6. How many leaves in the visualized tree label days as working days? Put your answer in the markdown cell in the end of the section.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(max_depth=4, random_state=42)
dt_model.fit(X_scaled, y)
```

```python
accuracy = dt_model.score(X_scaled, y)
accuracy
```

```python
for depth in range(2, 11):
    dt = DecisionTreeClassifier(max_depth=depth, random_state=42)
    dt.fit(X_scaled, y)
    acc = dt.score(X_scaled, y)
    print(f"Max Depth: {depth}, Accuracy: {acc}")
```

```python
xx, yy = np.meshgrid(
    np.arange(X_scaled[:, 0].min() - 0.5, X_scaled[:, 0].max() + 0.5, 0.1),
    np.arange(X_scaled[:, 1].min() - 0.5, X_scaled[:, 1].max() + 0.5, 0.1)
)
grid = np.c_[xx.ravel(), yy.ravel()]
Z = dt_model.predict(grid).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap="coolwarm", edgecolor="k")
plt.title("Decision Tree (max_depth=4)")
plt.xlabel("am (scaled)")
plt.ylabel("pm (scaled)")
plt.colorbar(label="Class")
plt.show()
```
