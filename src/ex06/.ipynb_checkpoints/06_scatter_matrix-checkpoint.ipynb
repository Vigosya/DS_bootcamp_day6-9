```python
import pandas as pd
import sqlite3
from pandas.plotting import scatter_matrix
```

```python
connection = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
pd.io.sql.read_sql("SELECT * FROM checker LIMIT 10;", connection)
```

```python
query = """
CREATE TABLE datamart AS
SELECT
    c.uid,
    c.labname,
    MIN(c.timestamp) AS first_commit_ts,
    MIN(p.datetime) AS first_view_ts
FROM
    checker c
LEFT JOIN
    pageviews p
ON
    c.uid = p.uid
WHERE 
    c.status = 'ready'
    AND c.numTrials = 1
    AND c.labname IN ('laba04', 'laba04s', 'laba05', 'laba06', 'laba06s', 'project1')
    AND c.uid LIKE 'user_%'
GROUP BY
    c.uid, c.labname

"""
```

```python
datamart = pd.io.sql.read_sql("SELECT * FROM datamart", connection)
```

```python
datamart['first_commit_ts'] = pd.to_datetime(datamart['first_commit_ts'])
datamart['first_view_ts'] = pd.to_datetime(datamart['first_view_ts'])
```

```python
test = datamart[datamart['first_view_ts'].notnull()].copy()
```

```python
test.to_sql('test', connection, if_exists='replace', index=False)
```

```python
query = """
WITH average_diffs AS (
SELECT
    t.uid,
    AVG((d.deadlines - strftime('%s', t.first_commit_ts)) / 3600.0) AS avg_diff
FROM
    test t
JOIN
    deadlines d
ON
    t.labname = d.labs
WHERE
    t.first_view_ts IS NOT NULL
    AND t.labname != 'project1'
GROUP BY
    uid
),
user_pageviews AS (
SELECT
    uid,
    COUNT(*) as pageviews
FROM
    pageviews
WHERE
    uid LIKE 'user_%'
GROUP BY
    uid
),
user_commits AS (
SELECT
    uid,
    COUNT(*) AS num_commits
FROM
    checker
WHERE
    uid LIKE 'user_%'
    AND labname != 'project1'
GROUP BY
    uid
)
SELECT
    d.uid,
    p.pageviews,
    c.num_commits,
    d.avg_diff
FROM
    average_diffs d
LEFT JOIN
    user_pageviews p
ON
    d.uid = p.uid
LEFT JOIN
    user_commits c
ON
    d.uid = c.uid;
"""
```

```python
df = pd.io.sql.read_sql(query, connection)
```

```python
scatter_matrix(
    df[['avg_diff', 'pageviews', 'num_commits']],
    figsize=(15,8),
    color='red',
    diagonal='kde',
    s=200,
    hist_kwds={'linewidth': 3}
)
```

## Can we say that if a user has a low number of pageviews then they likely have a low number of commits?
Answer: **No**

## Can we say that if a user has a low number of pageviews then they likely have a small average difference between the first commit and the lab deadline?
Answer: **No**

## Can we say that there are many users with a low number of commits and a few with a high number of commits?
Answer: **No**

## Can we say that there are many users with a small average difference and a few with a large average difference?
Answer: **No**
