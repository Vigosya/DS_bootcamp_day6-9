```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
connection = sqlite3.connect("../data/checking-logs.sqlite")
```

```python
query = """
SELECT strftime('%H', timestamp) AS hour,
    CASE
        WHEN strftime('%w', timestamp) IN ('1', '2', '3', '4', '5') THEN 'working_day'
        ELSE 'weekend'
    END AS day_type
FROM 
    checker
WHERE
    uid LIKE 'user_%';
""" 
df = pd.io.sql.read_sql(query, connection)
df['hour'] = df['hour'].astype(int)
```

```python
workdays_lst = list(df[df['day_type'] == 'working_day']['hour'])
weekends_lst = list(df[df['day_type'] == 'weekend']['hour'])
```

```python
plt.figure(figsize=(15, 8))
plt.hist(
    workdays_lst, bins=24, color='skyblue', alpha=0.7, label='working days',
)
plt.hist(
    weekends_lst, bins=24, color='red', alpha=0.7, label='weekend'
)
plt.xlim([0, 24])
plt.legend()
```

## Are there hours when the total number of commits was higher on weekends than on working days?

Yes! Top-4 hours: 11, 13, 22, 23.
