```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/ab-test.csv')
```

```python
data_after = data[data['time'] == 'after']
data_before = data[data['time'] == 'before']
```

```python
data_after_prep = [data_after[data_after['group'] == 'control']['diff'], data_after[data_after['group'] == 'test']['diff']]
data_before_prep = [data_before[data_before['group'] == 'control']['diff'], data_before[data_before['group'] == 'test']['diff']]
```

```python

fig, axes = plt.subplots(1, 2, figsize=(15, 8), sharey=True)
fig.suptitle('Lag before labs deadlines. After and before the Newsfeed')

axes[0].boxplot(data_after_prep, boxprops=dict(color='green', linewidth=3), medianprops=dict(color='red', linewidth=2), flierprops=dict(marker='o', color='red'))
axes[0].set_title('after')
axes[0].set_xticklabels(['control', 'test'])
axes[0].grid(True, axis='y')
axes[0].grid(True, axis='x')


axes[1].boxplot(data_before_prep, boxprops=dict(color='green', linewidth=3), medianprops=dict(color='red', linewidth=2), flierprops=dict(marker='o', color='red'))
axes[1].set_title('before')
axes[1].set_xticklabels(['control', 'test'])
axes[1].grid(True, axis='y')
axes[1].grid(True, axis='x')
```
